```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # Choose any! We auto support RoPE Scaling internally!
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.

llama_model, llama_tokenizer = FastLanguageModel.from_pretrained(
    model_name = "Antonio27/llama3-8b-4-bit-for-sugar",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)

gemma_model, gemma_tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/gemma-7b-bnb-4bit",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)
```

```python
#Init the test variables

question_1="What makes a rainbow?"
question_2="What are stars made of?"
question_3="How do airplanes fly?"
question_4="Why do we need to exercise?"
question_5="How do computers work?"

alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
{}

### Input:
{}

### Response:
{}"""
```

```python
# change the padding tokenizer value
llama_tokenizer.add_special_tokens({"pad_token": "<|reserved_special_token_0|>"})
llama_model.config.pad_token_id = llama_tokenizer.pad_token_id # updating model config
llama_tokenizer.padding_side = 'right' # padding to right (otherwise SFTTrainer shows warning)


FastLanguageModel.for_inference(llama_model)
inputs = llama_tokenizer(
    [
        alpaca_prompt.format(
            f"Your main task is to answer children's questions or tell stories to children. When you use any complex or difficult words, please explain them using very simple words that a 3-year-old child can understand. After explaining the concept, you can introduce the complex word. Also, ensure that your answers and stories are suitable and engaging for a 3-year-old.\n\nQuestion: {question_1}",  # instruction
            "",  # input
            "",  # output - leave this blank for generation!
        )
    ], return_tensors="pt").to("cuda")

outputs = llama_model.generate(**inputs, max_new_tokens=512, use_cache=True)
decoded_outputs = llama_tokenizer.batch_decode(outputs)

response_text = decoded_outputs[0]
response_start = response_text.find("### Response:") + len("### Response:")
initial_response = response_text[response_start:].strip()

# Remove the "### Explanation:" part if it exists
explanation_start = initial_response.find("### Explanation:")
if explanation_start != -1:
    initial_response = initial_response[:explanation_start].strip()

# end_of_text_token = "<|end_of_text|>"
# if end_of_text_token in initial_response:
#     initial_response = initial_response.split(end_of_text_token)[0].strip()

# Adjusting response with gemma_model
gemma_tokenizer.add_special_tokens({"pad_token": "<|reserved_special_token_0|>"})
gemma_model.config.pad_token_id = gemma_tokenizer.pad_token_id # updating model config
gemma_tokenizer.padding_side = 'right' # padding to right (otherwise SFTTrainer shows warning)

FastLanguageModel.for_inference(gemma_model)

inputs = gemma_tokenizer(
    [
        alpaca_prompt.format(
            f"The given content is: {initial_response}. Your task is to modify the given content to make it suitable for a 5-year-old child. Use simple words and phrases that a 5-year-old can easily understand. Also, ensure to remove any repetitive content.",  # instruction
            "",  # input
            "",  # output - leave this blank for generation!
        )
    ], return_tensors="pt").to("cuda")

outputs = gemma_model.generate(**inputs, max_new_tokens=256, use_cache=True)
decoded_outputs = gemma_tokenizer.batch_decode(outputs)

response_text = decoded_outputs[0]
response_start = response_text.find("### Response:") + len("### Response:")
adjusted_response = response_text[response_start:].strip()

if end_of_text_token in adjusted_response:
    adjusted_response = adjusted_response.split(end_of_text_token)[0].strip()
# Remove the "### Score:" part if it exists
score_start = adjusted_response.find("### Score:")
if score_start != -1:
    adjusted_response = adjusted_response[:score_start].strip()
# Remove the "### Instruction:" part if it exists
instruction_start = adjusted_response.find("### Instruction:")
if instruction_start != -1:
    adjusted_response = adjusted_response[:instruction_start].strip()
# Remove the "### Output:" part if it exists
output_start = adjusted_response.find("### Output:")
if output_start != -1:
    adjusted_response = adjusted_response[:output_start].strip()
# Remove the "### Evaluation:" part if it exists
eval_start = adjusted_response.find("### Evaluation:")
if eval_start != -1:
    adjusted_response = adjusted_response[:eval_start].strip()

print("\n")
print(f"Question1:", question_1)
print("\n")
print(f"Initial Response1:", initial_response)
print("\n")
print(f"Adjusted Response1:", adjusted_response)
print("\n")
```
